#### hyperparameter tuning (Bayesian optimization)

```python
import lightgbm
from bayes_opt import BayesianOptimization
from lightgbm import LGBMClassifier
```

```python
def lgb_hyper(n_estimators,learning_rate,max_depth,max_bin,num_leaves):
    params={
        'boosting_type':'gbdt',
        'metric':'auc',
        'n_estimators':int(n_estimators),
        'learning_rate':learning_rate,
        'max_depth':int(max_depth),
        'num_leaves':int(num_leaves),
        'max_bin':int(max_bin),
        'random_state':40,
    }
    lgtrain=lightgbm.Dataset(X,y)
    cv_result=lightgbm.cv(params,lgtrain,1000,early_stopping_rounds=300,stratified=True,nfold=5)
    return cv_result ['auc-mean'][-1]
```

```python
def lgb_train(n_estimators,learning_rate,max_depth,max_bin,num_leaves):
    params={
        'boosting_type':'gbdt',
        'n_estimators':int(n_estimators),
        'learning_rate':learning_rate,
        'max_depth':int(max_depth),
        'max_bin':int(max_bin),
        'num_leaves':int(num_leaves),
        'random_state':40,
    }
    classifier=LGBMClassifier(**params)
    classifier.fit(X,y)
    test_pred=classifier.predict(test_dataset)
    return test_pred
```

```python
def param_tuning(init_points,num_iter,**args):
    lgbBO=BayesianOptimization(lgb_hyper,{
        'n_estimators':(200,5000),
        'learning_rate':(0,0.5),
        'max_depth':(5,15),
        'max_bin':(100,600),
        'num_leaves':(10,400),
    })
    lgbBO.maximize(init_points=init_points,n_iter=num_iter,**args)
    return lgbBO'''
```
